Combine first four columns into one variable using string concatenation syntax. Construct new DataFrames with this new structure.

```python
# example: pd.concat([df1['c'], df2['c']], axis=1, keys=['df1', 'df2'])
df1['COMBINED_LOC'] = df1['C/A'] + ',' + df1['UNIT'] + ',' + df1['SCP'] + ',' + df1['STATION']
df1new = pd.concat([df1['COMBINED_LOC'], df1.ix[:, 4:11]], axis=1)
df1new.head()
```

```python
df2['COMBINED_LOC'] = df2['C/A'] + ',' + df2['UNIT'] + ',' + df2['SCP'] + ',' + df2['STATION']
df2new = pd.concat([df2['COMBINED_LOC'], df2.ix[:, 4:11]], axis=1)
df2new.head()
```

```python
df3['COMBINED_LOC'] = df3['C/A'] + ',' + df3['UNIT'] + ',' + df3['SCP'] + ',' + df3['STATION']
df3new = pd.concat([df3['COMBINED_LOC'], df3.ix[:, 4:11]], axis=1)
df3new.head()
```

```python
df4['COMBINED_LOC'] = df4['C/A'] + ',' + df4['UNIT'] + ',' + df4['SCP'] + ',' + df4['STATION']
df4new = pd.concat([df4['COMBINED_LOC'], df4.ix[:, 4:11]], axis=1)
df4new.head()
```
